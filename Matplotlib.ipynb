```python
#%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3,4])
plt.show()
```

```python
x = np.arange(1,5,1)
y = x*2
y
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
```

```python
a = np.linspace(1,10, 20)
b = a**1.5
c = (3*(a**3)/10)
fig, ax = plt.subplots(figsize=(10,10))
plt.plot(a,b,c)
ax.set(title= "Simple plot",
       xlabel = "This is the x-axis",
       ylabel = "This is the y-axis" )
plt.show()
```

```python
#Scatter plot
fig, ax = plt.subplots(figsize=(5,5))
ax.scatter(a, np.exp(a))
plt.show()
```

```python
#Line plot
fig, ax = plt.subplots()
ax.plot(a, np.sin(a))
plt.show()
```

```python
#Plotting from a dictionary
foods = {"Rice": 35,
        "Beans": 28,
        "Drinks": 41}
fig, ax = plt.subplots()
ax.bar(foods.keys(), height= foods.values())
ax.set(title= "A simple barplot",
      xlabel = "Food items",
      ylabel = "Food prices (NGN)")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.barh(list(foods.keys()), list(foods.values()))
```

```python
s = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(s);
```

```python

fig, (ax1, ax2) = plt.subplots(nrows= 1, ncols=2, figsize = (10,5))
ax1.plot(s, s/2)
ax2.scatter(np.random.random(10), np.random.random(10))
# ax3.bar(foods.keys(), foods.values())
# ax4.hist(np.random.randn(1000))
plt.show()
```

```python
#Using subplots.

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows= 2, ncols=2, figsize = (10,5))
ax1.plot(s, s/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(foods.keys(), foods.values())
ax4.hist(np.random.randn(1000))
plt.show()
```

```python
heart_disease = pd.read_csv("heart.csv")
```

```python
heart_disease.head()
```

```python
heart_disease.plot.hist()
```

```python
heart_disease["age"].plot.hist(bins=50)
```

```python
heart_disease.plot.hist(subplots=True, figsize=(10,30));
```

```python
over_50 =heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
over_50.plot(kind="scatter",
            x= "age",
            y='chol',
            c = "target");
```

```python
fig, ax = plt.subplots(figsize= (10,5))
over_50.plot(kind="scatter",
            x= "age",
            y='chol',
            c = "target",
            ax = ax);
ax.set_xlim([45,100])
```

```python
fig, ax = plt.subplots(figsize=(10,6))

scatter = ax.scatter(x=over_50["age"],
                    y= over_50["chol"],
                    c = over_50["target"])

ax.set(title = "Heart Disease and Cholesterol levels",
      xlabel ="Age",
      ylabel = "Cholesterol")

ax.legend(*scatter.legend_elements(), title= "Target");
ax.axhline(over_50["chol"].mean(),
          linestyle = "--");
```
