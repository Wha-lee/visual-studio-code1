```python
import numpy as np
import pandas as pd
```

```python
happy_data = pd.read_csv("happyscore_income.csv")
```

```python
happy_data.sort_values("income_inequality")
mid_income = happy_data["median_income"]
income_ineqq = happy_data["income_inequality"]
#print(mid_income, income_ineqq)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(y= income_ineqq, x = mid_income, s = 100, alpha=0.7, c="indigo")
plt.xlabel("Median Income")
plt.ylabel("Income inequality")
```

   Filtering were done on median income and income inequlity using the codes mid_income = happy_data["median_income"] and          income_ineqq = happy_data["income_inequality"] respectively.
